```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
df = pd.read_csv('credits.csv')
```

```python
literal_eval(df.crew[643])
```

```python
def func(x, y):
    gender_count = {'male': 0, 'female': 0}
    for i in literal_eval(x):
        if i['gender'] == 2:
            gender_count['male'] +=1
        else:
            gender_count['female'] +=1
    for j in literal_eval(y):
        if j['gender'] == 2:
            gender_count['male'] +=1
        else:
            gender_count['female'] +=1
    try:
        ratio = (gender_count['female']) / ((gender_count['male'] + (gender_count['female'])))
    except ZeroDivisionError:
        ratio = float("NaN")
    return(ratio)
```

```python
def func2(x):
    female_director = 0
    for i in literal_eval(x):
        if i['job'] == 'Director':
            if i['gender'] == 2:
                female_director = 0
            else:
                female_director = 1
    return(female_director)
```

```python
def func3(x):
    female_writer = 0
    for i in literal_eval(x):
        if i['job'] == 'Director':
            if i['gender'] == 2:
                female_writer += 0
            else:
                female_writer += 1
    return(female_writer)
```

```python
df['female_director'] = df['crew'].apply(func2)
```

```python
df['female_writers'] = df['crew'].apply(func3)
```

```python
#df['female_ratio'] = df['cast'].apply(func)
df['female_ratio'] = df.apply(lambda x: func(x['cast'], x['crew']), axis=1)
```

```python
df.drop(['cast', 'crew', 'female_writers'], axis = 1).to_csv('crew_cast_ratio.csv')
```
